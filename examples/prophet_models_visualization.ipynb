```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import ticker
formatter = ticker.ScalarFormatter(useMathText=True)
formatter.set_scientific(True) 
formatter.set_powerlimits((-1,1)) 
params = {'text.usetex' : True,
          'font.size' : 12,
          'font.family' : 'lmodern',
          }
plt.rcParams.update(params)
```

# To visualize the equations used in prophet

```python
# Visualize linear trend model: g(t)=kt+m
def linear_trend(t, k, m):
    """
    Linear trend model: g(t) = kt + m
    :param t: time
    :param k: slope
    :param m: intercept
    :return: g(t)
    """
    return k * t + m

t = np.linspace(0, 10, 100)  # time from 0 to 10
k = 1  # slope
m = 5  # intercept
g = linear_trend(t, k, m)

plt.figure(figsize=(2.5, 3))
plt.plot(t, g, label='Linear Trend: g(t) = kt + m')
plt.xlabel(r'$t$')
plt.ylabel(r'$g(t)$')
plt.tight_layout(h_pad=0.5, w_pad=0.1)
plt.savefig('linear.png', dpi=300)
```

```python
# Visualize linear logistic growth model: g(t)=C/(1+e^(-k*(t-m)))
def logistic_growth(t, C, k, m):
    """
    Logistic growth model: g(t) = C / (1 + e^(-k * (t - m)))
    :param t: time
    :param C: carrying capacity
    :param k: growth rate
    :param m: midpoint
    :return: g(t)
    """
    return C / (1 + np.exp(-k * (t - m)))
C = 1  # carrying capacity
k = 1  # growth rate
m = 5  # midpoint
g = logistic_growth(t, C, k, m)

plt.figure(figsize=(2.5, 3))
plt.plot(t, g, label='Logistic Growth: g(t) = C / (1 + e^(-k * (t - m)))')
plt.xlabel(r'$t$')
plt.ylabel(r'$g(t)$')
plt.tight_layout(h_pad=0.5, w_pad=0.1)
plt.savefig('logistic.png', dpi=300)
```
