```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import gridspec
```

```python
def get_data_dist_coeff(n, target=0, n_converge=52):
    """
    f(1) = 1
    f(n) → target
    Reaches ~target around n ≈ n_converge
    """
    alpha = np.log(2) / (n_converge - 1)  # ensures f(52) ≈ 0.5
    return target + (1 - target) * np.exp(-alpha * (n - 1))
```

```python
x = np.linspace(1, 300, 100)
y = [get_data_dist_coeff(n, target=0, n_converge=52) for n in x]
```

```python
fig = plt.figure(figsize=(5, 1))
gs = gridspec.GridSpec(1, 1)
ax0 = plt.subplot(gs[0])

ax0.plot(x, y, label='Data Distribution Coefficient', color='blue')
ax0.set_xlabel('x')
ax0.set_ylabel('Coeff')
ax0.grid(True)
# ax0.legend(fontsize=12)
ax0.set_ylim(-0.05, 1.05)
```
